```python
import pandas as pd 
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset = dataset.astype(int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
#standardization (improve)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
#model creation
from sklearn.svm import SVR
#regressor=SVR(kernel="linear",C=2000)
#regressor=SVR(kernel="rbf",C=10) #rbf-Radial bias function
#regressor=SVR(kernel="poly",C=10) #kernel and c is hyper tunning parameter
regressor=SVR(kernel="sigmoid",C=10)
#regressor=SVR(kernel="precomputed")
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_  
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
